**Excel: Questions on Conditional Formatting**

*1. Rules*

1. What is conditional formatting in Excel? 

--> Conditional formatting in Excel is a feature that allows you to apply specific formatting to cells that meet certain criteria or conditions. For example, you can highlight cells with values greater than a certain number.

2. How do you apply a basic conditional formatting rule in Excel? 

--> Select the range of cells you want to format, go to the "Home" tab, click on "Conditional Formatting," and then choose "Highlight Cells Rules" or "Top/Bottom Rules" and define your criteria. 

3. Can you create custom conditional formatting rules? Provide an example. 

--> Yes, you can create custom rules using formulas. For example, to highlight cells with values greater than the average, you can use the formula =A1>AVERAGE($A$1:$A$10) in the "Use a formula to determine which cells to format" option. 

*2. Data Bars*

4. What are Data Bars in Excel conditional formatting? 

--> Data Bars are a type of conditional formatting that visually represents the value of a cell relative to other cells with colored bars inside the cells. The length of the bar indicates the value. 

5. How do you apply Data Bars to a range of cells? 

--> Select the range of cells, go to "Conditional Formatting," choose "Data Bars," and select the desired style. 

6. Can you customize Data Bars? How? 

--> Yes, you can customize Data Bars by choosing "More Rules" after selecting a style. You can adjust the minimum and maximum values, bar color, and whether to show bars only or both bars and values. 

*3. Color Scales*

7. What are Color Scales in Excel? 

--> Color Scales are a type of conditional formatting that applies a gradient of colors to a range of cells based on their values. For example, a green-yellow-red scale can represent high, medium, and low values. 

8. How do you apply a Color Scale to a range of cells? 

--> Select the range of cells, go to "Conditional Formatting," choose "Color Scales," and select the desired color scale. 

9. Can you create custom Color Scales? 

--> Yes, you can create custom Color Scales by choosing "More Rules" after selecting a style. You can set specific colors for minimum, midpoint, and maximum values.

*4. Icon Sets* 

10. What are Icon Sets in Excel? 

--> Icon Sets are a type of conditional formatting that uses different icons to represent the values in a range of cells. For example, you can use arrows to indicate whether values are increasing, decreasing, or unchanged. 

11. How do you apply Icon Sets to a range of cells? 

--> Select the range of cells, go to "Conditional Formatting," choose "Icon Sets," and select the desired icon set. 

12. Can you customize Icon Sets? How? 

--> Yes, you can customize Icon Sets by choosing "More Rules" after selecting a set. You can adjust the criteria for each icon, such as changing the value thresholds or the icons used.

**Python : Basic Questions on File Handling**

*1. Reading/Writing Text Files*

1. How do you open a file in Python for reading? 

--> Use the `open` function with the mode `'r'`. For example: 

```python 
file = open('example.txt', 'r') 
``` 

2. How do you read the contents of a file line by line? 

--> Use a `for` loop to iterate through the file object. For example: 

```python 
with open('example.txt', 'r') as file: 
for line in file: 
print(line) 
```

3. How do you write to a file in Python? 

--> Use the `open` function with the mode `'w'` and the `write` method. For example: 

```python 
with open('example.txt', 'w') as file: 
file.write('Hello, world!') 
``` 

*2. Reading/Writing CSV Files*

4. How do you read a CSV file in Python? 

--> Use the `csv` module and the `csv.reader` function. For example: 

```python 
import csv 
with open('example.csv', 'r') as file: 
reader = csv.reader(file) 
for row in reader:  
print(row) 
``` 

5. How do you write to a CSV file in Python? 

--> Use the `csv` module and the `csv.writer` function. For example: 

```python 
import csv 
with open('example.csv', 'w', newline='') as file: 
writer = csv.writer(file) 
writer.writerow(['name', 'age']) 
writer.writerow(['Alice', 30]) 
``` 

*3. Reading/Writing JSON Files*

6. How do you read a JSON file in Python? 

--> Use the `json` module and the `json.load` function. For example: 

```python 
import json 
with open('example.json', 'r') as file: 
data = json.load(file) 
print(data) 
```

7. How do you write to a JSON file in Python? 

--> Use the `json` module and the `json.dump` function. For example: 

```python 
import json 
data = {'name': 'Alice', 'age': 30} 
with open('example.json', 'w') as file: 
json.dump(data, file) 
``` 

*4. Reading/Writing Binary Files*

```python
with open('example.txt', 'a') as file:
file.write('Append this text') 
``` 

3. How do you read and write compressed files in Python? 

--> Use the `gzip` module for gzip files or the `zipfile` module for zip files. For example: 

```python 
import gzip 
with gzip.open('example.txt.gz', 'rt') as file: 
content = file.read() 
``` 

*8. Advanced Questions on Exception Handling*

1. How do you create a custom exception with additional attributes in Python? 

--> Define a custom exception class with an `__init__` method to initialize additional attributes. For example:  

```python 
class CustomError(Exception): 
def __init__(self, message, code): 
super().__init__(message) 
self.code = code 
``` 

2. How do you log exceptions to a file in Python? 

--> Use the `logging` module to log exceptions. For example: 

```python 
import logging 
logging.basicConfig(filename='app.log', level=logging.ERROR) 
try: 
result = 10 / 0 
except ZeroDivisionError as e: 
    logging.error('Exception occurred', exc_info=True)
``` 

3. How do you re-raise an exception in Python after catching it? 

--> Use the `raise` keyword without any arguments. For example: 

```python 
try: 
result = 10 / 0 
except ZeroDivisionError: 
print('Handling exception') 
raise 
``` 


**SQL: Basic Questions on Data Aggregation**

*1. COUNT*

1. What does the COUNT function do in SQL? 

--> The COUNT function returns the number of rows that match a specified condition. For example: 

```sql 
SELECT COUNT(*) FROM employees; 
``` 

2. How can you count the number of distinct values in a column? 

--> Use the COUNT function with the DISTINCT keyword. For example: 

```sql 
SELECT COUNT(DISTINCT department) FROM employees; 
``` 

*2. SUM* 

3. What does the SUM function do in SQL? 

--> The SUM function returns the total sum of a numeric column. For example: 

```sql 
SELECT SUM(salary) FROM employees; 
``` 

4. Can you use SUM with a WHERE clause? Provide an example. 

--> Yes, you can use SUM with a WHERE clause to filter the rows. For example: 

```sql 
SELECT SUM(salary) FROM employees WHERE department = 'Sales'; 
``` 

*3. AVG* 

5. What does the AVG function do in SQL? 

--> The AVG function returns the average value of a numeric column. For example: 

```sql 
SELECT AVG(salary) FROM employees;  
``` 

6. How would you calculate the average salary for each department? 

--> Use the AVG function with the GROUP BY clause. For example: 

```sql 
SELECT department, AVG(salary) FROM employees GROUP BY department; 
``` 

*4. MIN* 

7. What does the MIN function do in SQL? 

--> The MIN function returns the smallest value in a column. For example: 

```sql 
SELECT MIN(salary) FROM employees; 
```

8. How can you find the minimum salary in each department? 

--> Use the MIN function with the GROUP BY clause. For example: 

```sql
SELECT department, MIN(salary) FROM employees GROUP BY department; 
``` 

*5. MAX* 

9. What does the MAX function do in SQL? 

--> The MAX function returns the largest value in a column. For example: 

```sql 
SELECT MAX(salary) FROM employees; 
``` 

10. How would you find the maximum salary for each job title? 

--> Use the MAX function with the GROUP BY clause. For example:  

```sql 
SELECT job_title, MAX(salary) FROM employees GROUP BY job_title; 
``` 

*Advanced Questions on Data Aggregation* 

1. How can you calculate the total, average, minimum, and maximum salaries in a single query? 

--> Use the SUM, AVG, MIN, and MAX functions in the SELECT clause. For example: 

```sql 
SELECT SUM(salary) AS total_salary, AVG(salary) AS avg_salary, MIN(salary) AS min_salary, 
MAX(salary) AS max_salary FROM employees; 
``` 

2. How would you find the number of employees with salaries above the average salary? 

--> Use a subquery to calculate the average salary, then count the employees with salaries above this value. For example: 

```sql 
SELECT COUNT(*) FROM employees WHERE salary > (SELECT AVG(salary) FROM employees); 
``` 

3. Can you use aggregate functions in the WHERE clause? Explain. 

--> No, aggregate functions cannot be used in the WHERE clause. Instead, use the HAVING clause to 
filter groups based on aggregate values. For example: 

```sql 
SELECT department, AVG(salary) FROM employees GROUP BY department HAVING AVG(salary) > 
50000; 
``` 

*Basic Questions on Database Design* 

*6. Normalization* 

1. What is normalization in database design? 

--> Normalization is the process of organizing a database to reduce redundancy and improve data integrity by dividing the database into tables and defining relationships between them. 

2. Explain the first normal form (1NF). 

--> A table is in 1NF if all columns contain only atomic, indivisible values and each column contains values of a single type. For example: 

```sql 
CREATE TABLE employees ( 
id INT PRIMARY KEY, 
name VARCHAR(100), 
phone_number VARCHAR(15) 
); 
``` 

3. What is the second normal form (2NF)? 

--> A table is in 2NF if it is in 1NF and all non-key columns are fully dependent on the primary key. This often involves removing partial dependencies by creating additional tables. 

4. Explain the third normal form (3NF). 

--> A table is in 3NF if it is in 2NF and all non-key columns are not only dependent on the primary key but also independent of each other. This removes transitive dependencies.

*7. ER Diagrams* 

5. What is an ER diagram in database design? 

--> An ER (Entity-Relationship) diagram is a graphical representation of entities and their relationships in a database. It shows how tables (entities) are connected and interact with each other. 
 
6. What are the main components of an ER diagram? 

--> The main components are entities (represented by rectangles), attributes (represented by ovals), 
and relationships (represented by diamonds or lines).  

*8. Primary and Foreign Keys* 
 
7. What is a primary key in a database? 

--> A primary key is a column or a set of columns that uniquely identifies each row in a table. For example: 
```sql 
CREATE TABLE employees ( id INT PRIMARY KEY, name VARCHAR(100) ); 
``` 
 
8. What is a foreign key in a database? 

--> A foreign key is a column or a set of columns that establishes a link between the data in two tables. It references the primary key of another table. For example: 

```sql 
CREATE TABLE orders ( order_id INT PRIMARY KEY, employee_id INT, FOREIGN KEY (employee_id) REFERENCES employees(id) ); 
``` 
 
9. Explain the purpose of a foreign key constraint. 

--> A foreign key constraint ensures referential integrity by enforcing a link between the foreign key column and the primary key column it references, preventing invalid data entry. 
 
*9. Advanced Questions on Database Design* 

1. How would you design a database for a library management system? Outline the main tables and their relationships. 

--> Design involves creating tables like `Books`, `Authors`, `Members`, `Loans`, and establishing relationships. For example: 

```sql 
CREATE TABLE Books ( 
book_id INT PRIMARY KEY, 
title VARCHAR(255), 
author_id INT, 
FOREIGN KEY (author_id) REFERENCES Authors(author_id) 
); 

CREATE TABLE Authors ( 
author_id INT PRIMARY KEY, 
name VARCHAR(255) 
); 

CREATE TABLE Members ( 
member_id INT PRIMARY KEY, 
name VARCHAR(255) 
); 

CREATE TABLE Loans ( 
loan_id INT PRIMARY KEY, 
book_id INT, 
member_id INT, 
loan_date DATE, 
return_date DATE, 
FOREIGN KEY (book_id) REFERENCES Books(book_id), 
FOREIGN KEY (member_id) REFERENCES Members(member_id) 
); 

``` 
2. What is denormalization and when would you use it? 

--> Denormalization is the process of combining tables to reduce the number of joins and improve query performance. It is used when read performance is more critical than write performance, such as in data warehousing. 

3. How can you enforce data integrity in a database design?  

--> Data integrity can be enforced using constraints such as primary keys, foreign keys, unique constraints, check constraints, and triggers. For example:
 
```sql 
CREATE TABLE employees ( 
id INT PRIMARY KEY, 
name VARCHAR(100) NOT NULL, 
email VARCHAR(100) UNIQUE 
); 
```
