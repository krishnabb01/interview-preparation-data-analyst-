*EXCEL: Basic Questions on Text Functions*

1. LEFT 

Q. What does the LEFT function do in Excel? 

--> The LEFT function extracts a specified number of characters from the beginning of a text string. For example: 

=LEFT("Hello World", 5)  # Result: "Hello" 

Q. How would you use the LEFT function to get the first three characters of the text in cell A1? 

--> Use the formula: 
=LEFT(A1, 3) 
______________________________________

2. RIGHT

Q. What does the RIGHT function do in Excel? 

--> The RIGHT function extracts a specified number of characters from the end of a text string. For example: 

=RIGHT("Hello World", 5)  # Result: "World" 

Q. How would you use the RIGHT function to get the last four characters of the text in cell B1? 

--> Use the formula: 
 
= RIGHT(B1, 4) 
___________________________

3. MID 

Q. What does the MID function do in Excel? 

--> The MID function extracts a specified number of characters from a text string, starting at any position. For example: 

=MID("Hello World", 7, 5)  # Result: "World" 

Q. How would you use the MID function to extract 3 characters starting from the 2nd character of the text in cell C1? 

--> Use the formula: 
=MID(C1, 2, 3)

4. LEN

Q. What does the LEN function do in Excel? 

--> The LEN function returns the length of a text string, including spaces. For example: 

=LEN("Hello World")  # Result: 11 

Q. How would you use the LEN function to find the number of characters in cell D1? 

--> Use the formula: 
=LEN(D1) 
___________________________

5. TRIM 

Q. What does the TRIM function do in Excel? 
 
-->The TRIM function removes all extra spaces from a text string, leaving only single spaces between words. For example: 

=TRIM("  Hello   World  ")  # Result: "Hello World" 

Q. How would you use the TRIM function to clean up the text in cell E1? 

--> Use the formula: 
=TRIM(E1)

6. UPPER

--> The UPPER function converts all lowercase letters in a text string to uppercase. For example: 

=UPPER("Hello World")  # Result: "HELLO WORLD" 

Q. How would you use the UPPER function to convert the text in cell F1 to uppercase? 

--> Use the formula: 
=UPPER(F1)

*Basic Questions on List Comprehensions*

1. What is a list comprehension in Python and how do you write one? 

--> A list comprehension is a concise way to create lists. It consists of brackets containing an expression followed by a for clause. For example, '[x * 2 for x in range(10)]' creates a list of the first 10 even numbers.  

2. How would you create a list of squares from 1 to 10 using list comprehension? 

--> You can create a list of squares using the following list comprehension: '[x**2 for x in range(1, 11)]'. 

3. Can you filter items in a list comprehension? Provide an example. 

--> Yes, you can add a condition to filter items. For example, '[x for x in range(10) if x % 2 == 0]' creates a list of even numbers from 0 to 9. 

4. How would you use a list comprehension to convert a list of strings to uppercase? 

--> You can use a list comprehension with the 'upper()' method: '[s.upper() for s in ['a', 'b', 'c']]'. 

5. Can list comprehensions include nested loops? Provide an example. 

--> Yes, you can include nested loops. For example, '[(x, y) for x in range(3) for y in range(3)]' creates a list of tuples with all combinations of 'x' and 'y'. 
______________________________________

*Advanced Questions on List Comprehensions* 

1. How would you flatten a nested list using list comprehensions? 

--> You can flatten a nested list using a nested list comprehension: '[item for sublist in nested_list for item in sublist]'. 

2. Provide an example of a nested list comprehension with a condition. 

--> '[[x * y for y in range(5) if y % 2 == 0] for x in range(5) if x % 2 == 0]' creates a 3x3 table for even numbers only. 

3. How can you use list comprehensions to create a dictionary? Provide an example. 

--> You can create a dictionary using a dictionary comprehension: '{k: v for k, v in zip(keys, values)}'. 

4. How can you use list comprehensions to process elements conditionally and transform them at the same time? 

--> Add a condition and an expression for transformation: '[x * 2 for x in range(10) if x % 2 == 0]' doubles even numbers from 0 to 9. 

5. What are the advantages of using list comprehensions over traditional loops? 

--> List comprehensions are more concise, readable, and often faster because they are optimized for Python's interpreter.


2. What are positional and keyword arguments? Provide examples. 

--> Positional arguments are passed in order. Keyword arguments are passed by explicitly naming them. For example: 

```python 
def greet(name, greeting="Hello"): 
print(f"{greeting}, {name}") 
greet("Alice")  # Positional 
greet(greeting="Hi", name="Bob")  # Keyword 
``` 

3. How do you import a module in Python? 

--> Use the 'import' keyword. For example, 'import math'. 

4. What is the difference between 'import' and 'from ... import'? 

--> 'import' brings in the entire module. 'from ... import' brings in specific attributes or functions. For example: 

```python 
import math 
from math import sqrt 
``` 

5. How can you create an alias for a module when importing it? 

--> Use the 'as' keyword. For example, 'import numpy as np'. 
________________________________________________________________________

*Advanced Questions on Functions and Modules* 

1. What are lambda functions and how do you create one?

--> Lambda functions are anonymous functions defined using the 'lambda' keyword. For example: 

```python 
add = lambda x, y: x + y 
``` 

2. How can you pass a function as an argument to another function? 

--> Functions can be passed as arguments just like any other object. For example: 

```python 
def apply_function(func, value): 
return func(value) 
``` 

3. Explain the concept of closures in Python. 

--> Closures are functions that capture variables from their enclosing scope. For example: 

```python 
def outer_func(x): 
def inner_func(y): 
return x + y 
return inner_func 
``` 

4. What is a decorator and how do you use one? 

--> A decorator is a function that modifies the behavior of another function. For example: 

```python 
def my_decorator(func): 
def wrapper(): 
print("Something is happening before the function is called.") 
func() 
print("Something is happening after the function is called.") 
return wrapper 
``` 


5. How do you handle exceptions in Python functions? 

--> Use 'try', 'except', 'else', and 'finally' blocks. For example: def divide(a, b): 

```python 
try: 
return a / b 
except ZeroDivisionError: 
return "Cannot divide by zero" 
``` 

6. What are the different ways to import only specific functions or variables from a module? 

--> You can use 'from ... import'. For example:  

```python 
from math import sqrt, pi 
``` 

7. How do you reload a module in Python? 

--> Use the 'reload' function from the 'importlib' module. For example: 

```python 
import importlib 
importlib.reload(module_name)
``` 
___________________________________________

**#SQL**

*Basic Questions on Subqueries and Nested Queries* 

1. Single-Row Subqueries 

1. What is a subquery in SQL?  

--> A subquery is a query nested inside another query. It can return a single value or a set of values and is used to perform operations that cannot be achieved with a single query. 

2. How do you use a single-row subquery in a SELECT statement? 

--> Single-row subqueries return one row. They can be used in the SELECT, WHERE, or HAVING clauses. For example: 

```sql 
SELECT name, salary 
FROM employees 
WHERE salary = (SELECT MAX(salary) FROM employees); 
``` 
____________________________________________

2. Multi-Row Subqueries 

3. What is a multi-row subquery? 

--> A multi-row subquery returns multiple rows. It is often used with operators like IN, ANY, or ALL. For example: 

```sql 
SELECT name, department 
FROM employees 
WHERE department IN (SELECT department FROM departments WHERE location = 'New 
York'); 
``` 

4. How do you use the IN operator with a multi-row subquery? 

--> The IN operator is used to filter rows based on a list of values returned by a subquery. For example: 

```sql 
SELECT name, department 
FROM employees 
WHERE department IN (SELECT department FROM departments WHERE location = 'New 
York'); 
``` 
___________________________________________

3. Correlated Subqueries 

5. What is a correlated subquery? 

--> A correlated subquery is a subquery that references columns from the outer query. It is executed once for each row processed by the outer query. For example: 

```sql 
SELECT e1.name, e1.salary 
FROM employees e1 
WHERE e1.salary > (SELECT AVG(e2.salary) FROM employees e2 
WHERE e1.department_id = e2.department_id); 
``` 

6. How do correlated subqueries differ from regular subqueries? 

--> Regular subqueries are executed once and their result is used by the outer query. Correlated subqueries are executed repeatedly, once for each row of the outer query, using values from the outer query.
_______________________________________________

4. Aggregate Functions 

7. What is the purpose of the GROUP BY clause in SQL? 

--> The GROUP BY clause is used to group rows that have the same values in specified columns into summary rows, like calculating the aggregate functions for each group. For example: 

```sql 
SELECT department, COUNT(*) 
FROM employees 
GROUP BY department; 
``` 

8. How do you use aggregate functions with the GROUP BY clause? 

-->Aggregate functions like COUNT, SUM, AVG, MAX, and MIN are used to perform calculations on each group of rows. For example:

```sql 
SELECT department, AVG(salary) 
FROM employees 
GROUP BY department; 
``` 
________________________________________________

5. Filtering Groups 

9. What is the HAVING clause and how is it different from the WHERE clause? 

--> The HAVING clause is used to filter groups created by the GROUP BY clause based on aggregate functions. The WHERE clause filters individual rows before grouping. For example: 

```sql 
SELECT department, AVG(salary) 
FROM employees 
GROUP BY department 
HAVING AVG(salary) > 50000; 
```

10. Can you use the HAVING clause without the GROUP BY clause?

--> Yes, you can use the HAVING clause without the GROUP BY clause when an aggregate function is used without grouping. For example: 

```sql 
SELECT COUNT(*) 
FROM employees 
HAVING COUNT(*) > 100; 
```

*Advanced Questions on Subqueries and Nested Queries*

1. How can you use a subquery in the FROM clause of a SELECT statement? 

--> Subqueries can be used in the FROM clause to create a derived table that the outer query can select from. For example: 

```sql 
SELECT dept_summary.department, dept_summary.total_salary 
FROM (SELECT department, SUM(salary) AS total_salary 
FROM employees 
GROUP BY department) AS dept_summary; 
``` 

2. What is the difference between a scalar subquery and a row subquery? 

--> A scalar subquery returns a single value and can be used where a single value is expected. A row subquery returns a single row with multiple columns and is used in situations where multiple columns are needed. 

3. Explain how to use EXISTS with a subquery. 

--> The EXISTS operator is used to check if a subquery returns any rows. It returns true if the subquery returns one or more rows. For example: 

```sql 
SELECT name 
FROM employees 
WHERE EXISTS (SELECT 1 FROM departments WHERE departments.id = 
employees.department_id); 
``` 

4. How can you perform a correlated update using a subquery? 

--> A correlated subquery can be used in an UPDATE statement to update rows based on a condition from another table. For example: 

```sql 
UPDATE employees e1 
SET salary = salary * 1.1 
WHERE EXISTS (SELECT 1 FROM departments d WHERE d.id = e1.department_id AND 
d.location = 'New York'); 
```

*Advanced Questions on GROUP BY and HAVING*

1. How can you group data by multiple columns in SQL? 

--> You can group data by multiple columns by listing them in the GROUP BY clause, separated by commas. For example: 

```sql 
SELECT department, job_title, COUNT(*) 
FROM employees 
GROUP BY department, job_title; 
``` 

2. How do you use the ROLLUP operator with GROUP BY? 

--> The ROLLUP operator is used to create subtotals and grand totals for a set of columns. For example: 

```sql 
SELECT department, job_title, COUNT(*) 
FROM employees 
GROUP BY ROLLUP (department, job_title); 
``` 

3. Explain how to filter groups using aggregate functions in the HAVING clause. 

--> The HAVING clause can be used to filter groups based on the result of aggregate functions. For example: 

```sql 
SELECT department, AVG(salary) 
FROM employees 
GROUP BY department 
HAVING AVG(salary) > 50000; 
``` 

4. How do you handle NULL values when using GROUP BY and aggregate functions? 

--> NULL values are treated as a distinct group in GROUP BY. Aggregate functions generally ignore NULL values, except for COUNT(*). You can use functions like COALESCE to handle NULL values. For example: 

```sql 
SELECT department, COUNT(COALESCE(salary, 0)) 
FROM employees 
GROUP BY department; 
```
____________________________

**#Power BI** 

*Basic Questions on DAX Basics* 

1. Calculated Columns 

1. What is a calculated column in Power BI and how do you create one? 

--> Calculated columns are DAX expressions that create new columns in a table. They are evaluated for each row in the table and can be created using the 'New Column' button in the modeling tab. For example, to create a column that calculates the total price from quantity and unit price:  

TotalPrice = Sales[Quantity] * Sales[UnitPrice] 

2. How does a calculated column differ from a measure in Power BI? 

--> A calculated column is computed row by row and stored in the table, while a measure is computed on the fly during aggregation and not stored in the table. Measures are usually used in the values area of visualizations. 

3. Q: Can you give an example of when to use a calculated column? 

--> Use a calculated column when you need to create new data based on existing data in each row. For example, concatenating first and last names: 

FullName = Customers[FirstName] & " " & Customers[LastName] 
_________________________

2. Measures 

4. What is a measure in Power BI and how do you create one? 

--> Measures are DAX expressions used to calculate values dynamically based on user interactions with the report. They are created using the 'New Measure' button in the modeling tab. For example, to create a measure that sums the total sales: 

TotalSales = SUM(Sales[SalesAmount]) 

5. How does context affect measures in Power BI? 

--> Measures are evaluated based on the filter context, which is determined by the data model and user interactions (such as slicers, filters, and visualizations). The same measure can produce different results depending on the context in which it is used. 

6. Provide an example of a measure using an aggregation function. 

--> A measure calculating the average sales per transaction: 

AverageSales = AVERAGE(Sales[SalesAmount]) 
___________________________

3. Aggregations 

7. What are aggregation functions in DAX and why are they important? 

--> Aggregation functions perform calculations on a set of values and return a single value. They are important for summarizing data in meaningful ways, such as sums, averages, counts, minimums, and maximums. 

8. Give an example of a DAX formula that uses the 'SUM' function.

--> To calculate the total revenue: 

TotalRevenue = SUM(Sales[Revenue])

 
9. How can you use the 'COUNT' function in DAX? Provide an example. 

--> To count the number of sales transactions: 

NumberOfSales = COUNT(Sales[TransactionID]) 
_____________________

4. Time Intelligence 

10. What is time intelligence in DAX and why is it useful? 

--> Time intelligence functions in DAX simplify calculations over date ranges, such as year-todate, quarter-to-date, and month-over-month comparisons. They are useful for analyzing trends and performance over time.  

11. Provide an example of a time intelligence function in DAX. 

--> To calculate year-to-date sales: 

YTDSales = TOTALYTD(SUM(Sales[SalesAmount]), Calendar[Date]) 


12. How would you calculate the previous year's sales using DAX? 

--> Use the 'SAMEPERIODLASTYEAR' function: 

PreviousYearSales = CALCULATE(SUM(Sales[SalesAmount]), 
SAMEPERIODLASTYEAR(Calendar[Date])) 
____________________

5. Filters 

13. How do you apply filters in DAX formulas? 

--> Filters can be applied using functions like 'CALCULATE', 'FILTER', and others to modify the filter context. For example: 

SalesInUSA = CALCULATE(SUM(Sales[SalesAmount]), Sales[Country] = "USA") 

14. What is the purpose of the 'CALCULATE' function in DAX? 

--> The 'CALCULATE' function changes the context in which the data is evaluated. It is used to apply filters to calculations, enabling complex aggregations and conditional logic. 

15. Give an example of a DAX formula that uses the 'FILTER' function. - To calculate total sales for a specific product category: 

SalesForCategory = CALCULATE(SUM(Sales[SalesAmount]), FILTER(Sales, Sales[Category] = 
"Electronics")) 
___________________
6. Contexts 

16. What are the different types of context in DAX? 

--> The two main types of context in DAX are row context and filter context. Row context refers to the current row in a table, while filter context refers to the set of filters applied to the data model. 

17. How does filter context affect the calculation of measures? 

--> Filter context determines which data is included in the calculation of a measure. It is influenced by slicers, filters, and the placement of fields in the report, allowing dynamic and interactive reporting. 

18. Provide an example where both row context and filter context are used in a DAX formula.

--> To calculate the percentage of total sales for each product: 

SalesPercentage =  
DIVIDE( 
Sales[SalesAmount],  
CALCULATE(SUM(Sales[SalesAmount]), ALL(Sales))
_________
